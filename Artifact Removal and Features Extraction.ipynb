# Artifact Removal

Identify and remove artifacts that can contaminate the EEG signals. to appyly this we run processing  Visualization Inspection, Automated Artifact Detection with ICA, and Manual Rejection

```python
# Wave channel eeg for processing articfact removal
freq_bands = ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']
channels = ['TP9', 'AF7', 'AF8', 'TP10']

# Sampling standart freq
sfreq = 256  # Hz

# check timestamp in eeg dat
if 'TimeStamp' in eeg_data.columns:

    # Convert 'TimeStamp' column to datetime format
    eeg_data['TimeStamp'] = pd.to_datetime(eeg_data['TimeStamp'], format='%Y-%m-%d %H:%M:%S.%f')

    # Calculate `mean data` from first Timestamp and convert to UTC
    start_time = eeg_data['TimeStamp'].iloc[0].to_pydatetime()

    # Convert datetime to UTC using timezone from datetime module
    start_time_utc = start_time.replace(tzinfo=timezone.utc)
    print(f"Start time (meas_date) dalam UTC: {start_time_utc}")
else:
    # If there is no Timestamp column, specify the start_time manually.
    start_time_utc = None
    print("Kolom 'TimeStamp' tidak ditemukan, menetapkan start_time ke None.")

# Loop through each frequency band and apply ICA
for band in freq_bands:
    # Select a channel that matches a particular wave
    channel_names = [f'{band}_{ch}' for ch in channels]
    
    # Check if the desired channel is in the eeg data DataFrame
    print(f"\nMemproses band: {band}")
    print("Channel yang diakses:", channel_names)
    
    # Make sure the column is in the DataFrame eeg data
    if all([ch in eeg_data.columns for ch in channel_names]):
        data = eeg_data[channel_names].values.T  # Transpose data
        
        # Create MNE info
        ch_names = channels  # without '_band'
        ch_types = ['eeg'] * len(ch_names)  # Type EEG Channel
        
        # Create info object for MEN
        info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=ch_types)
        
        # Create Raw object for MEN
        raw_band = mne.io.RawArray(data, info)
        
        # Set the start time of the recording (meas_date) using the set_meas_date() method.
        if start_time_utc is not None:
            raw_band.set_meas_date(start_time_utc)  # Mengatur meas_date menggunakan datetime dalam UTC
        
        # Apply standard montage 1020
        montage = mne.channels.make_standard_montage('standard_1020')
        raw_band.set_montage(montage)
        
        # Run ICA for artifact detection
        print(f"Jalankan ICA untuk {band}")
        ica = ICA(n_components=4, random_state=97, max_iter=800)
        ica.fit(raw_band)
        
        # ICA component visualization for artifact identification
        ica.plot_components()
        
        # Manual rejection (Componen 0 and 1 as artefact)
        ica.exclude = [0, 1]
        
        # ICA for artifact removal
        raw_clean = ica.apply(raw_band)
        
        # ICA visualization
        raw_clean.plot(title=f'{band} band (cleaned)')
        
        # Debugiing
        print(f"ICA diterapkan untuk {band}. Jumlah komponen yang dikeluarkan: {ica.exclude}")
    else:
        print(f"Kolom untuk {band} tidak ditemukan di DataFrame.")
```

# Features Extraction

## Time Exraction

```python
# Extract time domain
def extract_time_domain_features(epoch_data, channel_names):
    features = {}
    for i, channel_name in enumerate(channel_names):
        # Ekstrak mean, variance, dan skewness untuk setiap channel
        features[f'{channel_name}_mean'] = np.mean(epoch_data[i, :])
        features[f'{channel_name}_variance'] = np.var(epoch_data[i, :])
        features[f'{channel_name}_skewness'] = skew(epoch_data[i, :])
    return features

# Function to calculate the average amplitude in a given time window
def extract_windowed_amplitude(epoch_data, channel_names, sfreq):
    features = {}
    # Example time windows -500ms to 0ms and 0ms to +50 0ms
    samples_before = int(0.5 * sfreq)  # 500ms = 0.5 seconds
    samples_after = int(0.5 * sfreq)

    # Average amplitude before keypress (-500ms to 0ms)
    before_window = epoch_data[:, :samples_before]
    after_window = epoch_data[:, -samples_after:]

    for i, channel_name in enumerate(channel_names):
        # Average amplitude in time window
        features[f'{channel_name}_amplitude_before'] = np.mean(before_window[i, :])
        features[f'{channel_name}_amplitude_after'] = np.mean(after_window[i, :])

    return features
```

## Extract feature frequency domain

```python
# Function to extract frequency domain features using FFT
def extract_frequency_domain_features(epoch_data, channel_names, sfreq):
    features = {}
    for i, channel_name in enumerate(channel_names):
        # Calculate FFT
        fft_values = np.abs(fft(epoch_data[i, :]))
        freqs = np.fft.fftfreq(len(fft_values), d=1/sfreq)

        # Store power in the desired frequency band (eg. Delta, Theta, etc.)
        delta_power = np.sum(fft_values[(freqs >= 1) & (freqs <= 4)])
        theta_power = np.sum(fft_values[(freqs >= 4) & (freqs <= 8)])
        alpha_power = np.sum(fft_values[(freqs >= 8) & (freqs <= 13)])
        beta_power = np.sum(fft_values[(freqs >= 13) & (freqs <= 30)])
        gamma_power = np.sum(fft_values[(freqs >= 30) & (freqs <= 40)])

        features[f'{channel_name}_delta_power'] = delta_power
        features[f'{channel_name}_theta_power'] = theta_power
        features[f'{channel_name}_alpha_power'] = alpha_power
        features[f'{channel_name}_beta_power'] = beta_power
        features[f'{channel_name}_gamma_power'] = gamma_power

    return features
```

## Apply function and labelling

```python
# Step 1: Extract EEG data from raw_clean after ICA has been applied
sfreq = raw_clean.info['sfreq']  # Sampling frequency (Hz)
eeg_data = raw_clean.get_data()  # Get EEG data from MNE RawArray

# Step 2: Ensure that 'Keypress_Time' is in datetime format and add UTC timezone if needed
local_data['Keypress_Time'] = pd.to_datetime(local_data['Keypress_Time'], errors='coerce')
local_data['Keypress_Time'] = local_data['Keypress_Time'].dt.tz_localize('UTC')  # Ensure 'Keypress_Time' is in UTC timezone

# Step 3: Retrieve the start recording time (meas_date) from raw_clean and ensure it has a timezone
meas_date = raw_clean.info['meas_date']
if meas_date.tzinfo is None:
    meas_date = meas_date.replace(tzinfo=pytz.utc)  # Ensure 'meas_date' is in UTC timezone
print(f"Start recording time (meas_date) in UTC: {meas_date}")

# Step 4: Calculate time difference (delta_time) between 'Keypress_Time' and the recording start time (meas_date)
local_data['delta_time'] = (local_data['Keypress_Time'] - meas_date).dt.total_seconds()

# Debug: View 'delta_time' results to ensure the conversion is correct
print(local_data[['Keypress_Time', 'delta_time']].head())

# Step 5: Ensure there are no NaN values after calculating delta_time
local_data = local_data.dropna(subset=['delta_time'])

# Step 6: Convert delta_time to sample index
keystroke_samples = (local_data['delta_time'].values * sfreq).astype(int)

# Debug: Check the result of keystroke_samples
print(f"EEG sample limit: {len(eeg_data[0])}")
print("Keystroke samples (check first 10 values):", keystroke_samples[:10])

# Step 7: Create labels for classification (0 for left, 1 for right)
labels = np.where(local_data['Chosen_Key'].str.lower() == 'p', 1, 0)  # Label based on keypress (use lower() for consistency)

# Step 8: Segment EEG data into epochs (200ms before to 500ms after keypress)
epochs_data = []
epoch_labels = []
tmin = -0.2  # 200ms before keypress
tmax = 0.5   # 500ms after keypress

# Calculate the number of samples before and after the keystroke
samples_before = int(tmin * sfreq)
samples_after = int(tmax * sfreq)

# Debug: View the number of keystroke samples and matching labels
print(f"Number of keystroke samples: {len(keystroke_samples)}, Number of labels: {len(labels)}")

# Step 9: Segment EEG data for each keystroke and assign labels
for sample, label in zip(keystroke_samples, labels):
    start_sample = sample + samples_before
    end_sample = sample + samples_after

    # Debug: Check sample index bounds and epoch length
    if start_sample >= 0 and end_sample < len(eeg_data[0]):
        print(f"Start sample: {start_sample}, End sample: {end_sample}")
        epoch = eeg_data[:, start_sample:end_sample]  # Segment epoch from EEG data
        epochs_data.append(epoch)
        epoch_labels.append(label)
    else:
        print(f"Sample out of bounds: Start={start_sample}, End={end_sample}")

# Step 10: Convert epochs_data and epoch_labels into numpy arrays
epochs_data = np.array(epochs_data)
epoch_labels = np.array(epoch_labels)

# Debug: Check the shape of the resulting epoch data and labels
print(f"Shape of epochs data: {epochs_data.shape}")  # (n_epochs, n_channels, n_samples_per_epoch)
print(f"Labels (check first 10 values): {epoch_labels[:10]}")

# Step 11: Get channel names from raw_clean
channel_names = raw_clean.info['ch_names']

# Step 12: Feature Extraction from Each Epoch
dataset = []
for epoch, label in zip(epochs_data, epoch_labels):
    # Time-domain feature extraction
    time_features = extract_time_domain_features(epoch, channel_names)
    
    # (Optional) Frequency-domain feature extraction
    freq_features = extract_frequency_domain_features(epoch, channel_names, sfreq)
    
    # Combine extracted features
    combined_features = {**time_features, **freq_features}
    
    # Add label
    combined_features['label'] = label
    
    # Save to dataset
    dataset.append(combined_features)

# Step 13: Convert dataset into a DataFrame for further analysis
df_features = pd.DataFrame(dataset)

# Step 14: View a few rows from the feature extraction results
print(df_features.head())

# Step 15: Check label distribution
print("Label distribution:")
print(df_features['label'].value_counts())
```
